# 01 · Data Extraction

**Objective:** Load the raw Amazon Electronics dataset, perform an initial inspection, and understand the structure of the data before cleaning.

**Data source:** `raw.csv` — scraped Amazon search-result pages for electronics products (≈ 38 MB, ~110 K rows).

```python
# ── Imports ──────────────────────────────────────────
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# ── Load raw data ───────────────────────────────────
RAW_PATH = '../data/raw.csv'
df_raw = pd.read_csv(RAW_PATH)
print(f'Loaded {RAW_PATH}  →  {df_raw.shape[0]:,} rows × {df_raw.shape[1]} columns')
```

```python
# ── Quick preview: first & last 5 rows ─────────────
display(df_raw.head())
display(df_raw.tail())
```

```python
# ── Schema overview (dtypes + non-null counts) ─────
df_raw.info(verbose=True, show_counts=True)
```

```python
# ── Missing-value audit ────────────────────────────
missing = (
    df_raw.isnull().sum()
    .rename('missing_count')
    .to_frame()
    .assign(pct=lambda d: (d['missing_count'] / len(df_raw) * 100).round(2))
    .sort_values('pct', ascending=False)
)
display(missing)
```

```python
# ── Duplicate-row count ────────────────────────────
n_dupes = df_raw.duplicated().sum()
print(f'Duplicate rows: {n_dupes:,}  ({n_dupes / len(df_raw) * 100:.2f}%)')
```

```python
# ── Summary statistics (numeric + categorical) ────
display(df_raw.describe(include='all').T)
```
